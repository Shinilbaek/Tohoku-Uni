```python
spots_osaka = ['Universal Studio', 'Osaka Castle', 'Harbor Village',
               'Shitennoji Temple', 'Osaka Aquarium', 'National Museum of Art']

group_1_spots = ['Universal Studio', 'Osaka Aquarium', 
                 'Osaka Castle', 'Shitennoji Temple']
group_2_spots = ['Osaka Castle', 'Harbor Village', 
                 'Shitennoji Temple', 'National Museum of Art']
group_3_spots = ['Universal Studio', 'Harbor Village', 'Osaka Aquarium']

all_spots = group_1_spots+group_2_spots+group_3_spots
logic_list = []
for num in range(len(spots_osaka)):
    logic_tour = all_spots.count(spots_osaka[num]) >= 2
    logic_list.append(logic_tour)

spots_visit = []
for num in range(len(spots_osaka)):
    if logic_list[num]:
        spots_visit.append(spots_osaka[num])

print(f'Spots to Visit: ')
for spot in spots_visit:
    print(f'- {spot}')
```
