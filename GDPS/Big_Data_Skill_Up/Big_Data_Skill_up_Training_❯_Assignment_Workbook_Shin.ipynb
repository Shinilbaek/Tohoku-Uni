We want to make a simple code to clean and split some flower data we collected. By referring to the provided variables, write a code that gives the expected output (use all the tools you have learned so far).

```python
title = "- Sample of Flower Observation -"
flower_name = 'IRIS SETOSA'
data_name = 'PETAL WIDTH'

data_part1 = ['Part1', 5.1, '?', 4.9, 4.6, '']
data_part2 = ['Part2', 4.7, 5, 5.4, 4.6, '?', 5, 4.4, 4.9, '']
data_part3 = ['Part3', '?', 4.8, 4.8, 4.3, 5.8, 5.7, 5.3, '']

# a. Writing Header
print(title.upper())
print(f'Flower Name     :{flower_name.lower()}')
print(f'Measurement Name:{data_name.title()}')

# b. Making copy of original data
import copy
data_origin = [data_part1,data_part2,data_part3]
data = copy.deepcopy(data_origin)

# c. Removing first and last column
for i in range(3):
    data[i].pop(0)
    data[i].pop(-1)

print(f"Removing first and last column:\n{data[0]}\n{data[1]}\n{data[2]}\n")

# d. Removing unknown measurements ('?')
for i in range(3):
    data[i].remove('?')

print(f"Removing unknown measurements ('?'):\n{data[0]}\n{data[1]}\n{data[2]}\n")

# E. Counting length of each data
print(f"Data Part 1 Count:  {len(data[0])} measurements.")
print(f"Data Part 2 Count:  {len(data[1])}  measurements.")
print(f"Data Part 3 Count:  {len(data[2])} measurements.")
print(f"Data Total Count :  {len(data[0])+len(data[1])+len(data[2])} measurements.\n")

# f. Calculating Average of Part 1
print(f"Data Part 1 Average (index 0, 1, 2):\n{(data[0][0]+data[0][1]+data[0][2])/3:.3f}cm\n")

# g. Combining all Data
data_list = data[0]+data[1]+data[2]
print("All Data Combined:")
print(f'{data_list}\n')

# h. Splitting the Data into 4 parts (every 4 element):
print("Splitting the Data into 4 parts:")
data_list4 = [[],[],[],[]]
for i in range(4):
    for j in range(4):
        data_list4[i].append(data_list[i+j])
for i in range(4):
    print(data_list4[i])

# i. Printing Original Data:
print("Original Data:")
for i in range(3):
    print(data_origin[i])
```

### <font size='6'>**Assignment Lesson 2**&nbsp;&nbsp;</font>
We want to further understand our data by implementing for loop and if statement. By referring to the provided variables, write a code that gives the expected output using all the tools you have learned so far. Some comments in the code is provided as a guideline.

```python
data_cleaned = [[5.1, 4.9, 4.6],
                [4.7, 5, 5.4, 4.6, 5, 4.4, 4.9],
                [4.8, 4.8, 4.3, 5.8, 5.7, 5.3]]

# a. Printing Cleaned Data
print("Cleaned Data:")
print(f'- Part 1: {data_cleaned[0]}')
print(f'- Part 2: {data_cleaned[1]}')
print(f'- Part 3: {data_cleaned[2]}')

# b. Adding all elements in data_cleaned (using for loop)
print("\nSummed Data and Average:")
data_sum = []
for i in range(3):
    data_sum.append(sum(data_cleaned[i]))

print(f'- Part 1: Sum: {data_sum[0]}  Average: {data_sum[0]/len(data_cleaned[0])}')
print(f'- Part 2: Sum: {data_sum[1]}  Average: {data_sum[1]/len(data_cleaned[1])}')
print(f'- Part 3: Sum: {data_sum[2]}  Average: {data_sum[2]/len(data_cleaned[2])}')

# c. Getting the Data Between 4.5 and 5
data_cleaned_mid = [[],[],[]]
for i in range(3):
    data_cleaned_mid[i] = [data for data in data_cleaned[i]
                          if (data>=4.5) and (data<=5)]

print("\nData between 4.5 and 5:")
for i in range(3):
    print(f'- Part  {i+1}: {data_cleaned_mid[i]}\n')

# d. Counting Outlier (value more than 5.6)
data_cleaned_outl = [[],[],[]]
for i in range(3):
    data_cleaned_outl[i] = [data for data in data_cleaned[i]
                           if data > 5.6]

print("\nChecking Number of Outlier")
for i in range(3):
    print(f'- Part  {i+1}: {len(data_cleaned_outl[i])}')

# e. Categorizing Size
# 'S' : < 4.75 cm
# 'M' : 4.75 cm - 5.25 cm
# 'L' : > 5.25 cm
data_size = [[],[],[]]
for i in range(3):
    for j in range(len(data_cleaned[i])):
        if data_cleaned[i][j] <4.75:
            data_size[i].append('S')
        elif data_cleaned[i][j] >=4.75 and data_cleaned[i][j]<=5.25:
            data_size[i].append('M')
        elif data_cleaned[i][j] > 5.25:
            data_size[i].append('L')
        else:
            print('Error')
    
    
data_size_all = data_size[0]+data_size[1]+data_size[2]
print("\nCategorizing Size (Combined):")
print(data_size_all)
# f. Counting Categorized Data
print("\nCounting Categorized Data")
print(f"- 'S': {data_size_all.count('S')}")
print(f"- 'M': {data_size_all.count('M')}")
print(f"- 'L': {data_size_all.count('L')}")
```

We want to combine our data with other type of measurements and see how to manage them. Using the provided variables, make a program that gives the expected output using all the tools you have learned so far. Some comments in the code is provided as a guideline.

```python
flower_interest = {"Iris-versicolour", "Iris-setosa", "Iris-virginica"}

properties = ("sepal length", "sepal width", 
              "petal length", "petal width", "class")

data_set = [(5.1, 3.5, 1.4, 0.2, "Iris-setosa"),
            (4.9, 3.0, 1.4, 0.2, "Iris-setosa"),
            (7.0, 3.2, 4.7, 1.4, "Iris-versicolour"),
            (7.0, 3.4, 4.9, 1.5, "Iris-versicolour"),
            (6.4, 3.2, 4.5, 1.5, "Iris-germanica"),
            (6.3, 3.0, 4.4, 1.2, "Iris-unguicularis"),
            (6.9, 3.1, 4.9, 1.5, "Iris-pallida"),
            (7.1, 3.2, 4.8, 1.5, "Iris-pallida")]

# a. Printing properties
print("Data Properties:")
print(properties)

# b. Printing numerical properties (first four) using slice operator
print("\nNumerical Properties:")
print(properties[:4])

# c. Printing categorical properties (last one) using index on tuple
print("\nCategorical Properties (last one):")
print(properties[-1])

# d. Saving the last element of each tuple in data_set as a set
print("\nFlower Set (Observed) from data_set")
flower_set = set()
for i in range(len(data_set)):
    flower_set.add(data_set[i][-1])

print(flower_set)

# d. Getting the difference between flower_observed and flower_interest
print("\nFlower Set Not Used (Not in flower_interest)")
diff_flower = flower_set.difference(flower_interest)
print(diff_flower)

# e. Combining properties (as key) 
#    and columns in data_set (as values) in a dictionary
print("\nComposing Data as Lists in Dictionary")
data_set_col = [[],[],[],[],[]]
for i in range(5):
    for j in range(len(data_set)):
        data_set_col[i].append(data_set[j][i])
        
data_dic = {}
for i in range(5):
    data_dic.update({properties[i]:data_set_col[i]})
print(data_dic.keys())
print(data_dic.values())


# f. Getting values with the key 'petal width' 
print("\nGetting 'petal width' from the dictionary:")
print(data_dic['petal width'])

# g. Making a list of new dictionaries (properties (as key) 
#    and element in data_set (as values))
print("\nSeparating Each Data as Dictionaries")
sep_dic = []

for i in range(8):
    item_dic={}
    for j in range(5):
        item_dic.update({properties[j]:data_set[i][j]})
    sep_dic.append(item_dic)    
print(sep_dic)

# h. Combining row number (as key with format 'Data #') and 
#    the list of dictionaries (as values) in a new dictionary
print("\nComposing Data as Indexed Dictionary")
row_keys = ['Data 0','Data 1','Data 2','Data 3','Data 4','Data 5','Data 6','Data 7']
new_dict = {}
for i in range(len(row_keys)):
    new_dict.update({row_keys[i]:sep_dic[i]})

print(new_dict.keys())
print(new_dict.values())

# i. Getting the values with the key 'Data 5' from the previous dictionary
print("\nGetting 'Data 5' from the dictionary:")
print(new_dict['Data 5'])
```

### <font size='6'>**Assignment Lesson 4**&nbsp;&nbsp;</font>
The data obtained we obtained are not clean. We will make some functions to clean and calculate some part of the data. Using the provided variables, make a program that gives the expected output using all the tools you have learned so far. Some comments in the code is provided as a guideline.

```python
data_set =\
"5.1;3.5;1.4;0.2;Iris-setosa\n\
4.9;3.0;1.4;0.2;Iris-setosa\n\
7.O;3.Z;4.7;1.4;Iris-versicolour\n\
7.0;3.4;#.9;!.5;Iris-versicolour\n\
6.4;3.2;4.5;?;Iris-germanica\n\
6.3;3.0;4.4;1.2;Iris-unguicularis\n\
6.9;3.I;4,9;1.5;Iris-pallida\n\
7.1;3.2;4,8;1.5;Iris-pallida"

properties = ("sepal length", "sepal width", 
              "petal length", "petal width", "class")

# a. Importing Regex and splitting each row in data_set
print("Splitting entries of data_set into a list")
import re
data_1 = re.split('\n',data_set)
print(data_1)

# b. Defining a function that convert string to float.
#    It will return the original data if cannot be converted.
def str_to_fl(x):
    try:
        return float(x)
    except:
        return x

# c. Cleaning the data by 
#    splitting on ";",        
#    replacing I to 1, Z to 2, and O to 0 for numerical properties, and
#    using the previously defined function to convert any numbers into float
data_2 = []
for i in range(len(data_1)):
    data_2.append(re.split(';',data_1[i]))

data_3 = []
for i in range(len(data_2)):
    data_i = []
    for j in range(len(data_2[i])-1):
        data_i.append(str_to_fl(data_2[i][j].replace('I','1').replace('Z','2').replace('O','0')))
    data_i.append(data_2[i][-1])
    data_3.append(data_i)

print("\nCleaning Data in the list")
print(data_3)

# e. Combining properties (as key) 
#    and columns in data_set (as values) in a dictionary (like Assignment 4e)
print("\nComposing Data as Lists in Dictionary")

data_dic={}
data_4 = []
for i in range(len(data_3[1])):
    x=[]
    for j in range(len(data_3)):
        x.append(data_3[j][i])
    data_4.append(x)

for i in range(len(properties)):
        data_dic.update({properties[i]:data_4[i]})
  
print(data_dic.keys())
print(data_dic.values())

# f. Making a lambda function to calculate sepal/petal area
#    f(x, y) = (2/3) * x * y
cal_spa = lambda x, y : (2/3)*x*y

# g. Calculating the sepal area and petal area using the previous lambda function:
#    sepal_area = (2/3) * sepal length * sepal width
#    petal_area = (2/3) * petal length * petal width
#    where it is not possible, return '?'.

def cal_sepal_area():
    x=[]
    for i in range(len(data_dic['sepal length'])):
        try:
            x.append(cal_spa(data_dic['sepal length'][i],data_dic['sepal width'][i]))
        except:
            x.append('?')
    return x

def cal_petal_area():
    y=[]
    for i in range(len(data_dic['petal length'])):
        try:
            y.append(cal_spa(data_dic['petal length'][i],data_dic['petal width'][i]))
        except:
            y.append('?')
    return y  

# h. Adding sepal_area and petal_area to the dictionary.
print("\nAdding 'sepal area' and 'petal area' to dictionary")
sepal_area=cal_sepal_area()
petal_area=cal_petal_area()
data_dic.update({'sepal area':sepal_area})
data_dic.update({'petal area':petal_area})
print(sepal_area)
print(petal_area)

# i. Using a lambda fuction, create a check to see if there is any missing values in sepal area and petal area
#    If complete return '', and if missing return ''
print("\nAdding 'complete' to dictionary")
if_miss = lambda x : ('' if x == '?' else '')
data_sepal_miss = []
for i in range(len(sepal_area)):
    data_sepal_miss.append(if_miss(sepal_area[i]))

data_petal_miss = []
for i in range(len(petal_area)):
    data_petal_miss.append(if_miss(petal_area[i]))

data_dic.update({'complete':data_petal_miss})
print(data_petal_miss)

# j. Unzipping the keys and values from the dictionary.
#    After obtaining the values from the dictionary, remember to transpose them
print("\nCreating properties_new and data_set_new from dictionary")
new_zip = zip(data_dic.keys(),data_dic.values())
properties_new,data_set_new = zip(*new_zip)
print(properties_new)
print(data_set_new)
```

### <font size='6'>**Assignment Lesson 5**&nbsp;&nbsp;</font>
We will get more numerical properties from the data. Using the provided variables, make a program that gives the expected output using all the tools you have learned so far. Some comments in the code is provided as a guideline.

```python
properties = ("sepal length", "sepal width", "petal length", "petal width")

data_setosa = [(5.1, 3.5, 1.4, 0.2),
               (4.9, 3.0, 1.4, 0.2),
               (4.8, 3.0, 1.3, 0.1)]

data_versicolour = [(7.0, 3.2, 4.7, 1.4),
                    (7.0, 3.4, 4.9, 1.5),
                    (6.8, 3.0, 4.5, 1.3),
                    (6.9, 3.0, 4.3, 1.4)]
              
# a. Importing numpy, then converting data_setosa and data_versicolor to ndarray
import numpy as np
setosa_arr = np.array(data_setosa)
versicolour_arr = np.array(data_versicolour)

# b. Printing the dimensions and shape of the two arrays.
print("Printing the data and their dimension and shape")
print(f'{setosa_arr}')
print(f'Dimension : {setosa_arr.ndim}')
print(f'Shape     : {setosa_arr.shape}\n')

print(versicolour_arr)
print(f'Dimension : {versicolour_arr.ndim}')
print(f'Shape     : {versicolour_arr.shape}\n')


# c. Making two small random arrays (random float from 0 to 0.1) 
#    with the same shape as data_setosa and data_versicolor
np.random.seed(999)
random_arr_1 = np.random.uniform(0,0.1,size=(setosa_arr.shape))
random_arr_2 = np.random.uniform(0,0.1,size=(versicolour_arr.shape))

# d. Add the random arrays value in (b) to each data set, respectively.
print(f"Data with noise (3 decimal places)")
noise_1 = np.add(setosa_arr,random_arr_1.round(3))
noise_2 = np.add(versicolour_arr,random_arr_2.round(3))
print(noise_1)
print('\n')
print(noise_2)

# e. Splitting the sepal data and petal data
sepal_setosa = np.transpose(np.transpose(noise_1)[:2])
petal_setosa = np.transpose(np.transpose(noise_1)[2:])
sepal_versicolour = np.transpose(np.transpose(noise_2)[:2])
petal_versicolour = np.transpose(np.transpose(noise_2)[2:])
print(f'Sepal Data: Iris-setosa\n{sepal_setosa}')

# f. Computing Mean and Standard Deviation of each column
print("\nSepal Data: Iris-setosa")
print(sepal_setosa)
print("Mean and Standard Deviation")
print(np.mean(sepal_setosa,axis=0))
print(np.std(sepal_setosa,axis=0))
print('\n')

print("\nPetal Data: Iris-setosa")
print(petal_setosa)
print("Mean and Standard Deviation")
print(petal_setosa)
print(np.mean(petal_setosa,axis=0))
print(np.std(petal_setosa,axis=0))
print('\n')

print("\nSepal Data: Iris-versicolour")
print(sepal_versicolour)
print("Mean and Standard Deviation")
print(np.mean(sepal_versicolour,axis=0))
print(np.std(sepal_versicolour,axis=0))
print('\n')

print("\nPetal Data: Iris-versicolour")
print(petal_versicolour)
print("Mean and Standard Deviation")
print(np.mean(petal_versicolour,axis=0))
print(np.std(petal_versicolour,axis=0))

# g. Combining petal data and sepal data
print("\nSepal Data (combined):")
sepal_com = np.vstack([sepal_setosa,sepal_versicolour])
print(sepal_com)
print("\nPetal Data (combined):")
petal_com = np.vstack([petal_setosa,petal_versicolour])
print(petal_com)
# h. Filtering Sepal Data 
#    (omit if sepal length > 7.05)
#    (omit if sepal width > 3.5)


sl = np.transpose(sepal_com)[0]
sw = np.transpose(sepal_com)[1]

sepal_filter_1 = (sl < 7.05)
sepal_filter_2 = (sw < 3.5)

filtered_sl = sl[sepal_filter_1]
filtered_sw = sw[sepal_filter_2]
print("\nFiltered Sepal Length:")
print(filtered_sl)
print("\nFiltered Sepal Width")
print(filtered_sw)

# i. Writing 'Outlier' if sepal width > 3.2 and 'Not Outlier' for anything else
print("\nSeparating Outlier and Not Outlier based on sepal width:")
sw_out = np.where(sw>3.2,'Not Outlier','Outlier')
sepal_out=[]
for i in range(7):
    sepal_out.append([f'Data{i+1}',sw_out[i]])
    
print(sepal_out)
```
